```python
import pandas as pd
from time import time
import matplotlib.pyplot as plt

df = pd.read_csv('salaries.csv')

df['Date_Of_Payment'] = pd.to_datetime(df['Date_Of_Payment'])

# Filter the data for Poland employees in 2016
poland_employees_2016 = df[(df['Country'] == 'Poland') & (df['Date_Of_Payment'].dt.year == 2016)]

# Group by month and calculate the total monthly salary expenditure
monthly_salary_expenditure = poland_employees_2016.groupby(poland_employees_2016['Date_Of_Payment'].dt.strftime('%B'))['Monthly_Salary'].sum()

# Define the order of months
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

# Create a bar chart with months in order
plt.figure(figsize=(10, 6))
monthly_salary_expenditure[month_order].plot(kind='bar', color='skyblue')
plt.xlabel('Month')
plt.ylabel('Total Salary Expenditure (PLN)')
plt.title('Monthly Salary Expenditure for Poland Employees in 2016')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the chart
plt.show()
```
